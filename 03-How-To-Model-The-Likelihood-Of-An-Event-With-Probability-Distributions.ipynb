# How To Model The Likelihood Of An Event With Probability Distributions

## Uniform Distribution

An example would be "rolling a fair die" -- Each outcome of 1 through 6 is equally likely

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# "Roll the die" - a value of 1 through 6 is possible and equally likely
values = np.random.uniform(1, 6, 100000)
plt.hist(values, 500)
plt.show()
```

## Normal / Gaussian

The Normal (Gaussian) distribution can be used for modeling adult heights, newborn baby weights, ball bearing diameters, etc.

Let's generate some random numbers with a normal distribution:

```python
import numpy as np
import matplotlib.pyplot as plt

# "mu" is the desired mean, "sigma" is the standard deviation:
mu = 0.0
sigma = 4.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 500)
plt.show()
```

```python
# Variance should be equal to the square of the standard deviation
values.var()
```

## Exponential Probability Density Function

How long until an event?

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Binomial Probability Mass Function

The Binomial Distribution models the number of successes in things that act like a coin flip -- each flip is independent and has the same probability of success.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

## Poisson Probability Mass Function

Example: What are the odds of getting exactly 325 customers per day, given that my store gets, on average, 300 customers per day?

```python
from scipy.stats import poisson
from scipy.stats import norm
import matplotlib.pyplot as plt

mu = 300
x = np.arange(200, 400, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
